```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Super_hero/charcters_stats.csv')
```

```python
df.head()
```

```python
df.size
```

```python
df['Alignment'].value_counts()
```

```python
# Extracting only good characters from datasset
good = df[df['Alignment']=='good']
```

```python
good.head()
```

```python
# Sorting values interms of speed
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
# Sorting values interms of power
good.sort_values(by=['Power'],ascending=False).head()
```

```python
max_power = good[good['Power']==100]
```

```python
max_power.shape
```

```python
max_power
```

```python
good_max_power = good.sort_values(by=['Total'],ascending=False).head()
```

```python
good_max_power
```

```python
# Plotting a barplot of name vs total power
plt.figure(figsize=(10,8))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5])
plt.show()
```
