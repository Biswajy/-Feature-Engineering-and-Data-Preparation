```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Record of data where a team won batting first
batting_first = df[df['win_by_runs']!= 0]
```

```python
batting_first.head(5)
```

```python
plt.figure(figsize=(13,8))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
# creating a bar plot of top 5 teams with most wins after batting first
plt.figure(figsize=(13,5))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),batting_first['winner'].value_counts()[0:5])
plt.show()
```

```python
batting_second = df[df['win_by_wickets']!=0]
```

```python
batting_second.head(5)
```

```python
plt.figure(figsize = (10,5))
plt.hist(batting_second['win_by_wickets'],bins=10)
plt.title('Distribution of runs')
plt.xlabel('wickets')
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(12,5))
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]))
```
