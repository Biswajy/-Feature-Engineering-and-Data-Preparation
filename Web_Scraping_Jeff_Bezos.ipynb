We are going to use BeautifulSoup to scrape a wikipedia website to find desired data and will try to grab images present in it


```python
import requests
import bs4
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Jeff_Bezos')
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
# Response 200 represents a successful connection
print(res)
```

```python
# Grabbing the title
soup.select('title')
```

```python
# Grabbing the paragraphs
soup.select('p')
```

```python
soup.select('p')[1]
```

```python
# Grabbing the first paragraph
soup.select('h1')
```

```python
soup.select('title')[0].getText()
```

```python
# Creating a list of table of contents
soup.select('.toctext')
```

```python
for item in soup.select('.toctext'):
  print(item.text)
```
